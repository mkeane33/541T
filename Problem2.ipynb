```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
df = pd.read_csv('hour_score.txt', delimiter = "\t")
x = df['hour/week'].tolist() 
y = df['final score'].tolist()
```

```python
df.plot(kind='scatter',x='hour/week',y='final score',color='red')
plt.show()
```

```python

b0, b1 = 0.0, 1.0

lr = 0.0001
error = []
iter_ = 2000

for iter in range(iter_):
    costB0,costB1, iter_cost = 0, 0, 0
    for i in range(len(x)):
        #hypothesis
        h = (b0 + b1*x[i])
        #cost associated with the hypothesis
        iter_cost += (y[i] - h)**2 
        
        partialB1 = (-2 * x[i]) * (y[i] - (b0 + b1*x[i]))
        partialB0 = -2 * (y[i] - (b0 + b1*x[i]))
            
            
        #increase costs  
        costB1 += partialB1
        costB0 += partialB0

        # calculate new coeffs
        b1 = b1 - lr * costB1
        b0 = b0 - lr * costB0
        
    # keep track of errors - for visualization purposes
    error.append(iter_cost)
```

```python
y_preds = []
for i in x:
    
    y_preds.append( b0 + b1*i)
print(b0,b1)
print(y_preds)

plt.scatter(x,y, s=256, alpha = .3)
plt.plot(sorted(x), sorted(y_preds), linewidth = 4)
plt.show()
```

```python
print(error)
plt.plot(np.arange(1,len(error)+1), error, linewidth=4, )
plt.show()
```
